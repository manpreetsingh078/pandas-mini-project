```python
#explanayion if min max , averge dont shown on describe 

# Convert to numeric (coerce errors to NaN)
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')

# Now get min, max, mean
min_pay = data['BasePay'].min()
max_pay = data['BasePay'].max()
avg_pay = data['BasePay'].mean()

print("Min:", min_pay)
print("Max:", max_pay)
print("Average:", avg_pay)
```
